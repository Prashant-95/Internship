### Q1 and Q2

```python
#Take the user input
searchKey = input("Please type the searchkey ")
#Start the Automated web browser
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.amazon.in/")
time.sleep(6)
searchProduct = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
searchProduct.click()
searchProduct.send_keys(searchKey)
searchButton = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div")
searchButton.click()
time.sleep(8)

#Get links to individual products in a Page
Links_List = []
for page in range(0,3):
    links = driver.find_elements(By.XPATH, '//div[contains(@class, "s-result-item s-asin")]//h2//a[contains(@class, "a-text-normal")]')
    for link in links:
        Links_List.append(link.get_attribute("href"))
    #click on next button
    next_Btn = driver.find_element(By.XPATH, "//a[@class = 's-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    next_Btn.click()
    time.sleep(10)

brandName_List = []
productName_List = []
price_List = []
returnExchange_List =[]
expDelivery_List = []
availability_List = []

#Go to individual list and scrap all the details
for link in Links_List:
    driver.get(link)
    time.sleep(5)
    
    #Brands
    try:
        brand = driver.find_element(By.XPATH, '//tr[@class = "a-spacing-small po-brand"]//td[@class = "a-span9"]')
        brandName_List.append(brand.text)    
    except NoSuchElementException:
        brandName_List.append("-")
        
     #product Name
    try:
        product = driver.find_element(By.XPATH, '//tr[@class = "a-spacing-small po-model_name"]//td[@class = "a-span9"]')
        productName_List.append(product.text)
    except NoSuchElementException:
        productName_List.append("-")
        
    #PriceList
    try:
        price = driver.find_element(By.XPATH, '//span[@class = "a-price-whole"]')
        price_List.append(price.text)
    except NoSuchElementException:
        price_List.append("-")
        
    #return
    try:
        returns = driver.find_element(By.XPATH, '//ol[@class = "a-carousel"]//li[@class = "a-carousel-card tw-scroll-carousel-element"][2]')
        returnExchange_List.append(returns.text)
    except NoSuchElementException:
        returnExchange_List.append("-")
        
    #expected delivery
    try:
        expectedDelivery = driver.find_element(By.XPATH, '//span[@data-csa-c-type="element"]')
        expDelivery_List.append(expectedDelivery.get_attribute('data-csa-c-delivery-time'))
    except NoSuchElementException:
        expDelivery_List.append("-")
    
    #Avaialability List
    try:
        availability = driver.find_element(By.XPATH, '//div[@id="availability"]//span[@class= "a-size-medium a-color-success"]')
        availability_List.append(availability_List.text)
    except NoSuchElementException:
        availability_List.append("-")
                                           
print(len(brandName_List), len(productName_List), len(price_List), len(returnExchange_List), len(expDelivery_List), len(availability_List))

df = pd.DataFrame({"Brand Name" : brandName_List, "Product Name" : productName_List, "Price" : price_List, "ReturnorExchange" : returnExchange_List, "Expected Delivery" : expDelivery_List, "Availability" : availability_List})
df.to_csv('details.csv')
```

### Q 4)

```python
#Take the user input
searchKey = input("Please type the searchkey ")
#Start the Automated web browser
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.flipkart.com/")
time.sleep(6)

#Type product in search bar
if(driver.find_element(By.XPATH, "/html/body/div[2]/div/div/button").is_displayed):
    driver.find_element(By.XPATH, "/html/body/div[2]/div/div/button").click()
searchProduct = driver.find_element(By.XPATH, '//input[@class = "_3704LK"]')
searchProduct.click()
searchProduct.send_keys(searchKey)

#click ENTER
actions = ActionChains(driver)
actions.send_keys(Keys.ENTER)
actions.perform()
time.sleep(8)

#Get links to individual products in a Page
productUrl_List = []

#Fetch product urls
product_urls = driver.find_elements(By.XPATH, '//a[@class = "_1fQZEK"]')
for link in product_urls:
    productUrl_List.append(link.get_attribute("href"))


brandName_List = []
smartphoneName_List = []
price_List = []
colour_List =[]
RAM_List = []
storage_List = []
primaryCamera_List = []
secondaryCamera_List = []
displaySize_List = []
batteryCapacity_List = []


#Go to individual list and scrap all the details
for link in productUrl_List:
    driver.get(link)
    time.sleep(8)
    
    #Brands
    try:
        brand = driver.find_element(By.XPATH, '//span[@class = "B_NuCI"]')
        brandName_List.append(brand.text.split("\n")[0])    
    except NoSuchElementException:
        brandName_List.append("-")
        
     #smartphone name
    try:
        smartPhoneName = driver.find_element(By.XPATH, '//div[@class = "_3k-BhJ"][1]//tr[@class = "_1s_Smc row"][3]//li[@class= "_21lJbe"]')
        smartphoneName_List.append(smartPhoneName.text)
    except NoSuchElementException:
        smartphoneName_List.append("-")
        
    #PriceList
    try:
        price = driver.find_element(By.XPATH, '//div[@class = "_30jeq3 _16Jk6d"]')
        price_List.append(price.text)
    except NoSuchElementException:
        price_List.append("-")
        
    #colourList
    try:
        color = driver.find_element(By.XPATH, '//div[@class = "_3k-BhJ"][1]//tr[@class = "_1s_Smc row"][4]//li[@class= "_21lJbe"]')
        colour_List.append(color.text)
    except NoSuchElementException:
        colour_List.append("-")
        
    #RAM and Storage List
    try:
        RAMandStorage = driver.find_element(By.XPATH, '//div[@class = "_2418kt"]//li[@class = "_21Ahn-"][1]')
        ramstorageText = RAMandStorage.text.split("|")
        RAM_List.append(ramstorageText[0])
        storage_List.append(ramstorageText[1])
    except NoSuchElementException:
        RAM_List.append("-")
        storage_List.append("-")
    
    #Primery and Secondary Camera List
    try:
        cameras = driver.find_element(By.XPATH, '//div[@class = "_2418kt"]//li[@class = "_21Ahn-"][3]')
        primaryandsecondary = cameras.text.split("|")
        if(len(primaryandsecondary) < 2):
            primaryCamera_List.append(primaryandsecondary[0])
            secondaryCamera_List.append("-")
        else:
            primaryCamera_List.append(primaryandsecondary[0])
            secondaryCamera_List.append(primaryandsecondary[1])
    except NoSuchElementException:
        primaryCamera_List.append("-")
        secondaryCamera_List.append("-")
        
    #display size list
    try:
        display = driver.find_element(By.XPATH, '//div[@class = "_2418kt"]//li[@class = "_21Ahn-"][2]')
        displaySize_List.append(display.text)
    except NoSuchElementException:
        displaySize_List.append("-")
        
    #Battery Capacity
    try:
        battery = driver.find_element(By.XPATH, '//div[@class = "_2418kt"]//li[@class = "_21Ahn-"][4]')
        batteryCapacity_List.append(battery.text)
    except NoSuchElementException:
        batteryCapacity_List.append("-")
        
df = pd.DataFrame({"Brand Name" : brandName_List, "SmartPhone Name" : smartphoneName_List, "Colour" : colour_List, "RAM" : RAM_List, "Storage" : storage_List, "Primary Camera" : primaryCamera_List,
                  "Secondary Camera" : secondaryCamera_List, "Display Size" : displaySize_List, "Battery Capacity" : batteryCapacity_List, "Price" : price_List, "Product URL" : productUrl_List})                                          
df.to_csv('mobiledetails.csv')
```

```python
#Start the Automated web browser
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.hostelworld.com/")

#Type in search box
searchBox = driver.find_element(By.XPATH, '//input[@id = "location-text-input-field"]')
searchBox.click()
searchBox.send_keys("Lon")
time.sleep(20)


#click ENTER
actions = ActionChains(driver)
actions.send_keys(Keys.ARROW_DOWN)
time.sleep(2)
actions.send_keys(Keys.ENTER)
actions.perform()
time.sleep(3)



#Click on lets go
letsgobutton = driver.find_element(By.XPATH, '//button[@id = "search-button"]')
letsgobutton.click()
time.sleep(30)

#Empty lists
hostelName_List = []
distanceFromCity_List = []
ratings_List = []
totalReviews_List = []
privatePrice_List = []
dormsPrice_List = []
facilities_List = []

#Extract details
items = WebDriverWait(driver, 60).until(EC.presence_of_all_elements_located((By.XPATH, '//div[@class = "property-card"]')))
for i in items:
    #HostelName
    try:
        hostel = i.find_element(By.XPATH, '//h2[@class = "title title-6"]//a')
        hostelName_List.append(hostel.text)
    except NoSuchElementException:
        hostelName_List.append("-")
    
    #Distance from city
    try:
        distancefromcity = i.find_element(By.XPATH, '//div[@class = "property-distance"]//span[@class = "distance-description"]')
        distanceFromCity_List.append(distancefromcity.text)
    except NoSuchElementException:
        distanceFromCity_List.append("-")
    
    #ratings List    
    try:
        ratings = i.find_element(By.XPATH, '//div[contains(@aria-label, "Rating:")]')
        ratings_List.append(ratings.text)
    except NoSuchElementException:
        ratings_List.append("-")
    
    #Total Reviews
    try:
        reviews = i.find_element(By.XPATH, '//div[@class = "reviews"]')
        reviews_List.append(reviews.text)
    except NoSuchElementException:
        reviews_List.append("-")
        
    #Print length
    print(len(hostelName_List), len(distanceFromCity_List), len(ratings_List), len(reviews_List))
```
