### Q 1)

```python
#Browse the website
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
time.sleep(6)

# Empty Lists
rank_List = []
name_List = []
artist_List = []
uploadDate_List = []
views_List = []

#Find the elements
#Rank
try:
    ranks = driver.find_elements(By.XPATH, '//table[@class = "wikitable sortable jquery-tablesorter"][1]//tbody//tr//td[1]')
    for rank in ranks:
        rank_List.append(rank.text)
except NoSuchElementException:
    for i in range(0, 30):
        rank_List.append("-")

#Video Name
try:
    videoNames = driver.find_elements(By.XPATH, '//table[@class = "wikitable sortable jquery-tablesorter"][1]//tbody//tr//td[2]')
    for videoName in videoNames:
        name_List.append(videoName.text.split('"')[1])
except NoSuchElementException:
    for i in range(0, 30):
        name_List.append("-")
        
#Uploader
try:
    uploaders = driver.find_elements(By.XPATH, '//table[@class = "wikitable sortable jquery-tablesorter"][1]//tbody//tr//td[3]')
    for uploader in uploaders:
        artist_List.append(uploader.text)
except NoSuchElementException:
    for i in range(0, 30):
        artist_List.append("-")

#Upload Date
try:
    uploadDates = driver.find_elements(By.XPATH, '//table[@class = "wikitable sortable jquery-tablesorter"][1]//tbody//tr//td[5]')
    for uploadDate in uploadDates:
        uploadDate_List.append(uploadDate.text)
except NoSuchElementException:
    for i in range(0, 30):
        uploadDate_List.append("-")
        
#Views
try:
    views = driver.find_elements(By.XPATH, '//table[@class = "wikitable sortable jquery-tablesorter"][1]//tbody//tr//td[4]')
    for view in views:
        views_List.append(view.text)
except NoSuchElementException:
    for i in range(0, 30):
        views_List.append("-")
        
df = pd.DataFrame({"Rank" : rank_List, "Video Name" : name_List, "Artist" : artist_List, "Upload Date" : uploadDate_List, "Views in billlion" : views_List})
print(df)
```

```python
#Browse the website
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.statisticstimes.com/")
time.sleep(6)

#click on economy
economy = driver.find_element(By.XPATH, '//div[@class ="dropdown"][2]')
economy.click()

#Select the india in economy section
india = driver.find_element(By.XPATH, '//div[@class ="dropdown"][2] //div[@class="dropdown-content"]//a[text() = "India"]')
india.click()
time.sleep(10)

#Click on GDP of Indian States
gdp = driver.find_element(By.XPATH, '//a[contains(text(), "GDP of Indian states")]')
gdp.click()
time.sleep(8)

#Empty Lists
rank_List = []
state_List = []
GSDP_1819_List = []
GSDP_1920_List = []
share_List = []
GDPbln_List = []

#Rank

ranks = driver.find_elements(By.XPATH, '//div[@id = "table_id_wrapper"]//tbody//td[@class = "data1"]')
for rank in ranks:
    rank_List.append(rank.text)

#State
states = driver.find_elements(By.XPATH, '//div[@id = "table_id_wrapper"]//tbody//td[@class = "name"]')
for state in states:
    state_List.append(state.text)
    
#GSDP_1819
GSDP_1819 = driver.find_elements(By.XPATH, '//div[@id = "table_id_wrapper"]//tbody//td[@class = "data"][1]')
for i in GSDP_1819:
    GSDP_1819_List.append(i.text)
    
#GSDP_1920
GSDP_1920 = driver.find_elements(By.XPATH, '//div[@id = "table_id_wrapper"]//tbody//td[@class = "data sorting_1"]')
for i in GSDP_1920:
    GSDP_1920_List.append(i.text)
    
#Share
shares = driver.find_elements(By.XPATH, '//div[@id = "table_id_wrapper"]//tbody//td[@class = "data"][2]')
for share in shares:
    share_List.append(share.text)
    
#GDP Billion
GDPs = driver.find_elements(By.XPATH, '//div[@id = "table_id_wrapper"]//tbody//td[@class = "data"][3]')
for gdp in GDPs:
    GDPbln_List.append(gdp.text)
    
#DataFrame
df = pd.DataFrame({"Rank" : rank_List, "States" : state_List, "GSDP_1819" : GSDP_1819_List, "GSDP_1920" : GSDP_1920_List, "Share" : share_List, "GDp in billion" : GDPbln_List})

print(df)
```

```python
#Browse the website
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://github.com/")
time.sleep(6)

#click on Open source
openSource = driver.find_element(By.XPATH, '//li//button[contains(text(), "Open Source")]')
openSource.click()

#Click on Trending
trending = driver.find_element(By.XPATH, '//li//a[contains(text(), "Trending")]')
trending.click()
time.sleep(10)

#Store repository addresses
links_List = []
links = driver.find_elements(By.XPATH, '//h2[@class ="h3 lh-condensed"]//a')
for link in links:
    links_List.append(link.get_attribute("href"))
    
title_List = []
description_List = []
contriCount_List = []
language_List = []

#scrap the details
for link in links_List:
    driver.get(link)
    time.sleep(6)
    
    #Title
    try:
        title = driver.find_element(By.XPATH, '//strong[@class = "mr-2 flex-self-stretch"]//a')
        title_List.append(title.text)
    except NoSuchElementException:
        title_List.append("-")
        
    #description
    try:
        descriptiont = driver.find_element(By.XPATH, '//p[@class = "f4 my-3"]')
        description_List.append(descriptiont.text)
    except NoSuchElementException:
        description_List.append("-")
        
    #contributors
    try:
        contributor = driver.find_element(By.XPATH, '//a[@class = "Link--primary no-underline"]//span[@class = "Counter"]')
        contriCount_List.append(contributor.text)
    except:
        contriCount_List.append("-")
        
    #Languages
    try:
        language = driver.find_element(By.XPATH, '//span[@class = "color-fg-default text-bold mr-1"][1]')
        language_List.append(language.text)
    except:
        language_List.append("-")
        
df = pd.DataFrame({"Repo Title" : title_List, "Descrtiption" : description_List, "Contributers Count" : contriCount_List, "Language" : language_List})
print(df)
```

```python
#Browse the website
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare")
time.sleep(6)

#Empty lists
bookName_List = []
authorName_List = []
volume_List = []
publisher_List = []
genre_List = []

#Scrap the details
#Name
books_Name = driver.find_elements(By.XPATH, '//table[@class = "in-article sortable"]//tbody//tr//td[1]')
for book in books_Name:
    bookName_List.append(book.text)
    
#author
authors = driver.find_elements(By.XPATH, '//table[@class = "in-article sortable"]//tbody//tr//td[2]')
for author in authors:
    authorName_List.append(author.text)
    
#volume
volumes = driver.find_elements(By.XPATH, '//table[@class = "in-article sortable"]//tbody//tr//td[3]')
for volume in volumes:
    volume_List.append(volume.text)
    
#Publisher
publishers = driver.find_elements(By.XPATH, '//table[@class = "in-article sortable"]//tbody//tr//td[4]')
for publisher in publishers:
    publisher_List.append(publisher.text)
    
#genre
genres = driver.find_elements(By.XPATH, '//table[@class = "in-article sortable"]//tbody//tr//td[5]')
for genre in genres:
    genre_List.append(genre.text)
    
df = pd.DataFrame({"Book Name" : bookName_List, "Author" : authorName_List, "Volume" : volume_List, "Publisher" : publisher_List, "Genre" : genre_List})
df
```

### Q 7)

```python
#Browse the website
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.imdb.com/list/ls095964455/")
time.sleep(6)

#Empty Lists
name_List = []
yearSpan_List = []
genre_List = []
runTime_List = []
ratings_List = []
voters_List = []

#get all series
items = WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.XPATH, '//div[@class = "lister-item mode-detail"]')))

#All details to be filled in list
for item in items:
    #Name
    try:
        name = item.find_element(By.XPATH, './/h3[@class = "lister-item-header"]//a')
        name_List.append(name.text)
    except NoSuchElementException:
        name_List.append("-")
    
    #Year
    try:
        year = item.find_element(By.XPATH, './/h3[@class = "lister-item-header"]//span[@class = "lister-item-year text-muted unbold"]')
        yearSpan_List.append(year.text)
    except NoSuchElementException:
        yearSpan_List.append("-")
        
    #genre
    try:
        genre = item.find_element(By.XPATH, './/span[@class = "genre"]')
        genre_List.append(genre.text)
    except NoSuchElementException:
        genre_List.append("-")
        
    #runtime
    try:
        runtime = item.find_element(By.XPATH, './/span[@class = "runtime"]')
        runTime_List.append(runtime.text)
    except NoSuchElementException:
        runTime_List.append("-")
        
    #ratings
    try:
        rating = item.find_element(By.XPATH, './/div[@class = "ipl-rating-star small"]//span[@class = "ipl-rating-star__rating"]')
        ratings_List.append(rating.text)
    except NoSuchElementException:
        ratings_List.append("-")
        
    #Voters
    try:
        voters = item.find_element(By.XPATH, './/span[@name = "nv"]')
        voters_List.append(voters.text)
    except NoSuchElementException:
        voters_List.append("-")
        
df = pd.DataFrame({"Name" : name_List, "Years Span" : yearSpan_List, "Genre" : genre_List, "Runtime" : runTime_List, "Ratings" : ratings_List, "Voters" : voters_List})
df
```

### Q8)

```python
#Browse the website
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://archive.ics.uci.edu/")
time.sleep(6)

#click on all data set
allDataset = driver.find_element(By.XPATH, '//span[@class = "whitetext"]//b[text()= "View ALL Data Sets"]')
allDataset.click()
time.sleep(8)

#Empty Lists
datasetName_List = []
datatype_List = []
task_List = []
attributestype_List = []
instances_List = []
attributes_List = []
years_List = []

#Scrap all rows
rows = driver.find_elements(By.XPATH, '//table[@cellpadding = "5" and @border = "1"]/tbody/tr')

#Scrap Data from each row
for row in rows:
    #Names
    try:
        name = row.find_element(By.XPATH, './td//p[@class = "normal"]//a')
        datasetName_List.append(name.text)
    except NoSuchElementException:
        datasetName_List.append("-")
        
    #Type
    try:
        datatype = row.find_element(By.XPATH, './td[2]/p[@class = "normal"]')
        datatype_List.append(datatype.text.split('&')[0])
    except NoSuchElementException:
        datatype_List.append("-")
        
    #task
    try:
        task = row.find_element(By.XPATH, './td[3]/p[@class = "normal"]')
        task_List.append(task.text.split('&')[0])
    except NoSuchElementException:
        task_List.append("-")
    
    #attributes type
    try:
        attribute = row.find_element(By.XPATH, './td[4]/p[@class = "normal"]')
        attributestype_List.append(attribute.text.split('&')[0])
    except NoSuchElementException:
        attributestype_List.append("-")
        
    #Instances
    try:
        instance = row.find_element(By.XPATH, './td[5]/p[@class = "normal"]')
        instances_List.append(instance.text.split('&')[0])
    except NoSuchElementException:
        instances_List.append("-")
        
    #attribute
    try:
        attribute = row.find_element(By.XPATH, './td[6]/p[@class = "normal"]')
        attributes_List.append(attribute.text.split('&')[0])
    except NoSuchElementException:
        attributes_List.append("-")
        
    #years
    try:
        year = row.find_element(By.XPATH, './td[7]/p[@class = "normal"]')
        years_List.append(year.text.split('&')[0])
    except NoSuchElementException:
        years_List.append("-")
        
df = pd.DataFrame({"DataSet Name" : datasetName_List, "Data Type" : datatype_List, "Task" : task_List, "Attribute Type" : attributestype_List, "Instance" : instances_List, "Attributes" : attributes_List, "Years" : years_List})
df      
    
```

### Q 9)

```python
#Browse the website
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.naukri.com/hr-recruiters-consultants")
time.sleep(6)

#Type in search bar
search = driver.find_element(By.XPATH, '//input[@name = "qp"]')
search.send_keys("Data Science")

#click on search
btnSearch = driver.find_element(By.XPATH, '//button[@id = "qsbFormBtn"]')
btnSearch.click()
time.sleep(6)

#Empty Lists
name_List = []
designation_List = []
company_List = []
skills_List = []
location_List = []

#scrap the recruites
recruiters = driver.find_elements(By.XPATH, '//div[@class = "recInfo"]')

#Scrap all details 1 by 1
for recruiter in recruiters:
    #Name
    try:
        name = recruiter.find_element(By.XPATH, './/span[@class = "fl ellipsis"]')
        name_List.append(name.text)
    except NoSuchElementException:
        name_List.append("-")
        
    #Designation
    try:
        designation = recruiter.find_element(By.XPATH, './/span[@class = "ellipsis clr"]')
        designation_List.append(designation.text)
    except NoSuchElementException:
        designation_List.append("-")
        
    #comapny
    try:
        company = recruiter.find_element(By.XPATH, './/a[@class = "ellipsis"][2]')
        company_List.append(company.text)
    except NoSuchElementException:
        company_List.append("-")
        
    #Skills
    try:
        skill = recruiter.find_element(By.XPATH, './/div[@class = "hireSec highlightable"]')
        skills_List.append(skill.text)
    except NoSuchElementException:
        skills_List.append("-")
        
    #location
    try:
        location = recruiter.find_element(By.XPATH, './/small[@class  = "ellipsis"]')
        location_List.append(location.text)
    except NoSuchElementException:
        location_List.append("-")
    
df = pd.DataFrame({"Name" : name_List, "Designation" : designation_List, "Comapny" : company_List, "Skills" : skills_List, "Location" : location_List})
df
```
