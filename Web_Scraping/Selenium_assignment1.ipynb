### Q1)

```python
#Start the Automated web browser
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.naukri.com/")

#Pass the words Data Analyst and Bangalore to web
designation = driver.find_element(By.CLASS_NAME, "suggestor-input ")
designation.send_keys("Data Analyst")
location = driver.find_element(By.XPATH, "/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys("Bangalore")

#click on button "Search"
btnSearch = driver.find_element(By.CLASS_NAME, "qsbSubmit")
btnSearch.click()

#scrap the web, create and print Data Frame
jobTitle_List = []
jobLocation_List = []
companyName_List = []
experience_List = []

jobTitle = driver.find_elements(By.XPATH, '//a[@class="title ellipsis"]')
for job in jobTitle[0:10]:
    jobTitle_List.append(job.text)

jobLocation = driver.find_elements(By.XPATH, '//span[@class = "ellipsis fleft locWdth"]')
for location in jobLocation[0:10]:
    jobLocation_List.append(location.text)
    
companyName = driver.find_elements(By.XPATH, '//a[@class = "subTitle ellipsis fleft"]')
for company in companyName[0:10]:
    companyName_List.append(company.text)
    
experiences = driver.find_elements(By.XPATH,'//span[@class = "ellipsis fleft expwdth"]')
for experience in experiences[0:10]:
    experience_List.append(experience.text)

#Checking the Length
print(len(jobTitle_List), len(jobLocation_List), len(companyName_List), len(experience_List))

df = pd.DataFrame({"Job_Title" : jobTitle_List, "Job_Location" : jobLocation_List, "Company_Name" : companyName_List, "Experience" : experience_List})
print(df)
```

```python
#Start the Automated web browser
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.flipkart.com/")

#wait configuartion
wait = WebDriverWait(driver, 30)

#Closing of login window
time.sleep(6)
if(driver.find_element(By.XPATH, "/html/body/div[2]/div/div/button").is_displayed):
    driver.find_element(By.XPATH, "/html/body/div[2]/div/div/button").click()

#Pass the words Data Scientist web
searchProduct = driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input")
searchProduct.click()
searchProduct.send_keys("sunglasses")
searchButton = driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
searchButton.click()

#Scrap the data
time.sleep(5)
start = 0
end = 3
#Filter only Flipkart assured so that attributes can be scraped easily
wait.until(EC.visibility_of_element_located((By.XPATH, '//div[@class = "_3U-Vxu"]')))
driver.find_element(By.XPATH, '//div[@class = "_3U-Vxu"]').click()
time.sleep(5)

#define empty lists to store values
brands_List = []
descriptions_List = []
price_List = []

for page in range(start,end):
    brands = driver.find_elements(By.CLASS_NAME, '_2WkVRV')
    for i in brands:
        brands_List.append(i.text)

    descriptions = driver.find_elements(By.XPATH, '//a[@class = "IRpwTa"]')
    for i in descriptions:
        descriptions_List.append(i.text)
        
    prices = driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')
    for i in prices:
        price_List.append(i.text)
    
    nextButton = driver.find_element(By.XPATH, '//span[text() = "Next"]')
    nextButton.click()
    time.sleep(6)

print(len(brands_List), len(descriptions_List), len(price_List))

#Create and print data frame for first 100 elements
df = pd.DataFrame({"Brands" : brands_List[0:100], "Description" : descriptions_List[0:100], "Price" : price_List[0:100]})
print(df.all)
```

```python
#Start the Automated web browser
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.amazon.in/")

#wait configuartion
wait = WebDriverWait(driver, 30)

#Pass the words Data Scientist web
time.sleep(6)
searchProduct = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
searchProduct.click()
searchProduct.send_keys("Laptop")
searchButton = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div")
searchButton.click()
time.sleep(8)

#Check for CPU processor
driver.find_element(By.XPATH, '//span[text() = "Intel Core i7"]').click()


#define empty lists to store values
titles_List = []
ratings_List = []
price_List = []

items = WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.XPATH, '//div[contains(@class, "s-result-item s-asin")]')))

for item in items[0:10]:
    #Titles
    titles_amazon = item.find_element(By.XPATH, './/span[@class = "a-size-medium a-color-base a-text-normal"]')
    titles_List.append(titles_amazon.text)

    #ratings
    ratings_box = item.find_elements(By.XPATH, './/div[@class="a-row a-size-small"]/span')
    if ratings_box != []:
        ratings_Points = ratings_box[0].get_attribute('aria-label')
        ratings_Num = ratings_box[1].get_attribute('aria-label')
    else:
        ratings_Points, ratings_Num = 0, 0
    ratings_List.append(ratings_Points)
    
    #Prices
    prices = item.find_element(By.XPATH, './/div[@class = "sg-col-inner"]/div/div[@class = "a-row a-size-base a-color-base"]//span[@class = "a-price-whole"]')
    price_List.append(prices.text)
    

print(len(titles_List), len(ratings_List), len(price_List))

#Create and print data frame for first 10 elements
df = pd.DataFrame({"Titles" : titles_List, "Ratings" : ratings_List, "Price" : price_List})
print(df)
```

### Q 8)

```python
#Start automated web scrapper
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.azquotes.com/")
time.sleep(8)

#Click on Top 1000 button
btn_Top1000 = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
btn_Top1000.click()

#Empty Lists
quotes_List = []
authors_List = []
quotesType_List = []

#Scrap elements
for page in range(0,10):
    quotes = driver.find_elements(By.XPATH, '//div[@class = "wrap-block"]//a[@class = "title"]')
    for quote in quotes:
        quotes_List.append(quote.text)

    authors = driver.find_elements(By.XPATH, '//div[@class = "wrap-block"]//div[@class = "author"]/a')
    for author in authors:
        authors_List.append(author.text)

    quotesTypes = driver.find_elements(By.XPATH, '//div[@class = "wrap-block"]//div[@class = "tags"]')    
    for types in quotesTypes:
        quotesType_List.append(types.text)
    
    if(page != 9):
        btn_Next = driver.find_element(By.XPATH, '//li[@class = "next"]')
        btn_Next.click()
        time.sleep(5)
    
print(len(quotes_List), len(authors_List), len(quotesType_List))

#create data frame
df = pd.DataFrame({"Quotes" : quotes_List, "Authors" : authors_List, "Quotes Type" : quotesType_List})
print(df)
```

```python
#Start automated web browser
driver = webdriver.Chrome(r"E:\Internship\chromedriver.exe")
driver.get("https://www.jagranjosh.com/")
time.sleep(6)

#click on accept button
accept = driver.find_element(By.XPATH, '//a[@class = "closegdper"]')
accept.click()

#Click on GK Tab
tab_GK = driver.find_element(By.XPATH, '//div[@class = "hashtag"]//a[text() = "GK"]')
tab_GK.click()
time.sleep(6)

#click on list of all prime minister
PMs_List = driver.find_element(By.XPATH, '//div[@id = "popluarGK"]//*[contains(text(), "List of all Prime Ministers")]')
PMs_List.click()
time.sleep(6)

#Lists to store element values
PMsNames_List = []
Born_dead_List = []
TermofOffice_List = []
Remarks_List = []

#Names of PM's
PMs_Name = driver.find_elements(By.XPATH, '//div[@class = "table-box"]//td[2]//p')
for name in PMs_Name:
    PMsNames_List.append(name.text)

Born_Dead = driver.find_elements(By.XPATH, '//div[@class = "table-box"]//td[3]//p')
for i in Born_Dead:
    Born_dead_List.append(i.text)
    
termofOffice = driver.find_elements(By.XPATH, '//div[@class = "table-box"]//td[4]/p[1]')
for i in termofOffice: 
    TermofOffice_List.append(i.text)
    
remarks = driver.find_elements(By.XPATH, '//div[@class = "table-box"]//td[5]//p')
for i in remarks:
    Remarks_List.append(i.text)

print(len(PMsNames_List), len(Born_dead_List), len(TermofOffice_List), len(Remarks_List))

#create Data Frame
df = pd.DataFrame({"Names" : PMsNames_List, "Born_Dead" : Born_dead_List, "Term of office" : TermofOffice_List, "Remarks" : Remarks_List})
print(df)
```
